```python
#Mutation AA
def delete_other_mut(df_Cosmic):
    df_Cosmic = df_Cosmic.loc[~df_Cosmic['Mutation AA'].str.contains('delins')]
    #df_Cosmic = df_Cosmic.loc[~df_Cosmic['Mutation AA'].str.contains('Sec')]
    df_Cosmic = df_Cosmic.loc[~df_Cosmic['Mutation AA'].str.contains('fs')]
    def format_selenocysteine(x):
        if x[2:5] != ('Sec'):
            return x
        elif x[2:5] == ('Sec'):
            return x.replace(x[2:4] , 'U')
    df_Cosmic['Mutation AA'] = df_Cosmic['Mutation AA'].apply(format_selenocysteine)
    return df_Cosmic
```

```python
#add new column
def filter_on_index(df_Cosmic,col_name):
    indice_Cosmic = pd.read_csv('Filtered Cosmic Index.csv')
    indice_Cosmic.set_index([col_name], drop=False, append=False, inplace=True, verify_integrity=True)
    return indice_Cosmic.merge(df_Cosmic, how='inner', left_index=True, right_index=True, validate='1:1')
```

```python
#Accession Number, Transcript stable ID version(mappit file)
def add_uniprotid(Cosmic,mappit):
    Cosmic = Cosmic.rename(columns={'Accession Number': 'Trans version'})
    mappit = mappit.rename(columns={'Transcript stable ID version': 'Trans version'})
    return Cosmic.merge(mappit, how='inner', on='Trans version', sort=False, indicator=False, copy=False, validate='m:m' )
```

```python
cosmic = add_uniprotid(cosmic,mappit)
```

```python
#Mutation AA, ACC number
#takes long time to run
def domain_mapping(Cosmic,pfam_phpsites):
    
    Cosmic = Cosmic.rename(columns={'UniProtKB/Swiss-Prot ID': 'Acc ID'})
    pfam_phpsites = pfam_phpsites.rename(columns={'ACC_ID': 'Acc ID'})
    
    n = 1000  #chunk row size
    list_df = [Cosmic[i:i+n] for i in range(0,Cosmic.shape[0],n)]
    for idx,df in enumerate(list_df):
        df = df.merge(pfam_phpsites, how='inner', on='Acc ID', sort=False, indicator=False, copy=False, validate='m:m' )
        df['Mutation AA_sliced'] = df['Mutation AA'].apply(lambda x: to_int(x))
        df['Distance'] = df['UP_POS'] - df['Mutation AA_sliced']
        df = df[df['Distance'] <= 5]
        df = df[df['Distance'] >= -5]
        list_df[idx] = df
        
    return pd.concat(list_df)
```

```python
cosmic = domain_mapping(cosmic,pfam_mapped_phpsites)
```

```python
# mutation AA, UP_POS, Modification_name 
#only run after domain_mapping
def match_position_AA(df):
    df['Mutation AA_sliced'] = df['Mutation AA'].apply(lambda x: int(x[3:-1]))
    df = df.loc[df['Mutation AA_sliced'] == df['UP_POS']]
    df['modified AA'] = [x[2] for x in list(df['Mutation AA'])]
    df['Modification_name_trial'] = [x[-1] for x in list(df['Modification_name'])]
    df =  df.loc[df['modified AA'] == df['Modification_name_trial']]
    return df.drop(['Modification_name_trial','modified AA','Mutation AA_sliced'],axis=1)
```

```python
#UP_POS, UNP_START, UNP_END
#only run after domain_mapping
def check_domain_in_boundary(df):
    return df[df.apply(lambda x: (x['UP_POS'] >= x['seq_start']) & (x['UP_POS']<= x['seq_end']), axis=1)]
```

```python
#UP_POS, UNP_START, UNP_END
#only run after domain_mapping
def check_domain_out_boundary(df):
    return df[~df.apply(lambda x: (x['UP_POS'] >= x['seq_start']) & (x['UP_POS']<= x['seq_end']), axis=1)]
```
